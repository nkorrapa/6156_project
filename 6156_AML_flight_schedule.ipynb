```python
df_flights['CRS_DEP_TIME'] = df_flights['CRS_DEP_TIME'].astype(str)
df_flights['DEP_TIME'] = df_flights['DEP_TIME'].astype(str)
df_flights['WHEELS_OFF'] = df_flights['WHEELS_OFF'].astype(str)
df_flights['WHEELS_ON'] = df_flights['WHEELS_ON'].astype(str)
df_flights['CRS_ARR_TIME'] = df_flights['CRS_ARR_TIME'].astype(str)
df_flights['ARR_TIME'] = df_flights['ARR_TIME'].astype(str)
```
